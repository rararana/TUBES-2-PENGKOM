```python
data = pd.read_csv("Steel_industry_data.csv")
numericdata = ["Usage_kWh", "Lagging_Current_Reactive.Power_kVarh", "Leading_Current_Reactive_Power_kVarh", "Lagging_Current_Power_Factor", "Leading_Current_Power_Factor", "NSM"]
for i in range(len(numericdata)):
    olah = numericdata[i]
    print(str(i+1) + ". " + olah)
    print("rata rata: " + str(data[olah].mean()))
    print("persentil: ")
    print("\t10%: " + str(data[olah].quantile(0.10)))
    print("\t25%: " + str(data[olah].quantile(0.25)))
    print("\t50%: " + str(data[olah].quantile(0.50)))
    print("\t75%: " + str(data[olah].quantile(0.75)))
    print("\t90%: " + str(data[olah].quantile(0.90)))
    print("nilai ekstrem: ")
    print("\tmax: " + str(data[olah].max()))
    print("\tmin: " + str(data[olah].min()))
```

KESIMPULAN
rata rata penggunaan listrik adalah 27 kwh. hal ini cukup miris karena seperti yang bisa dilihat dari data persentilnya, 50% dari pengguna bahkan tidak mencapai 5 kwh. hal ini mengindikasikan bahwa adanya kesenjangan sosial yang sangat besar antara pengguna listrik. bahkan, penggunaan listrik 10% pengguna teratas hampir mencapai 80 kwh. penggunaan listrik mereka berkali kali lipat dibanding 50% penguna lainnya.
